```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import pickle

import sys
sys.path.append('/Users/shinbo/PycharmProjects/paper/LDA')
def print_top_words(lam, feature_names, n_top_words):
    for topic_id, topic in enumerate(lam):
        print('\nTopic Nr.%d:' % int(topic_id + 1))
        print(''.join([feature_names[i] + ' ' + str(round(topic[i], 2))
                       + ' | ' for i in topic.argsort()[:-n_top_words - 1:-1]]))

dir_ = '/Users/shinbo/Desktop/metting/LDA/0. data/JH_data/JH_data_cleaned.pkl'
data = pickle.load(open(dir_, 'rb'))
stop_words = ['good','great','go','get','place']
for i in range(len(data)):
    data[i] = [w for w in data[i] if w not in stop_words]
data_join = [' '.join(doc) for doc in data]
cv = CountVectorizer()
X = cv.fit_transform(data_join).toarray()
```

```python
dir_model_4 = '/Users/shinbo/Desktop/metting/LDA/0. data/JH_data/result_4_topics.pkl'
dir_model_5 = '/Users/shinbo/Desktop/metting/LDA/0. data/JH_data/result_5_topics.pkl'

def make_top_n_words(dir_, n):
    DMM_LDA = pickle.load(open(dir_,'rb'))
    K = DMM_LDA.components_.shape[0]
    result = pd.DataFrame()
    for k in range(K):
        top_words = pd.DataFrame({'word':list(cv.get_feature_names()), 'degree':DMM_LDA.components_[k,:]})
        top_words = top_words.sort_values(by='degree',ascending=False).iloc[:n,:]['word'].tolist()
        result[k] = top_words
    result.columns = ['Topic ' + str(i) for i in range(1,K+1)]
    return result
n = 50
result_4 = make_top_n_words(dir_model_4,n)
result_5 = make_top_n_words(dir_model_5,n)
```
